```python
from graphviz import Digraph

# 다이어그램 생성 (초등 1~6학년 "변화와 관계"/"자료와 가능성")
dot = Digraph(comment="초등 수학 - 변화와 관계, 자료와 가능성", format="png", engine="dot")
```

```python
# 노드 추가
# 2. 변화와 관계
dot.node("02", "변화와 관계")  # 대분류


dot.node("02-01", "규칙 찾기")  # 중분류
dot.node("02-02", "규칙 정하기")
dot.node("02-03", "등호와 동치 관계")
dot.node("02-04", "대응 관계")
dot.node("02-05", "비와 비율")
dot.node("02-06", "비례식과 비례배분")


dot.node("02-01-01", "배열에서 규칙 찾기")  # 소분류
dot.node("02-02-01", "규칙에 따라 배열하기")
dot.node("02-03-01", "등호 사용하기")
dot.node("02-04-01", "대응 관계를 나타낸 표에서 규칙 찾기") # ...
dot.node("02-05-01", "두 양의 크기 비교하기")
dot.node("02-06-01", "비례식 이해하기")
dot.node("02-06-02", "비례배분 이해하기")


dot.node("02-01-01-01", "물체 배열 규칙을 여러 가지 방법으로 표현하기")  # 하위분류
dot.node("02-01-01-02", "무늬 배열 규칙을 여러 가지 방법으로 표현하기")
dot.node("02-01-01-03", "수 배열 규칙을 여러 가지 방법으로 표현하기")
dot.node("02-01-01-04", "다양한 변화 규칙을 수로 나타내기")
dot.node("02-01-01-05", "다양한 변화 규칙을 식으로 나타내기")
dot.node("02-01-01-06", "계산식 배열 규칙으로 계산 결과 추측하기")

dot.node("02-02-01-01", "물체 배열하기")
dot.node("02-02-01-02", "무늬 배열하기")
dot.node("02-02-01-03", "수 배열하기")

dot.node("02-03-01-01", "크기가 같은 두 양의 관계를 식으로 나타내기")

dot.node("02-04-01-01", "□, △ 를 사용하여 식으로 나타내기")

dot.node("02-05-01-01", "두 양의 관계를 비로 나타내기")
dot.node("02-05-01-02", "비율을 분수로 나타내기")
dot.node("02-05-01-03", "비율을 소수로 나타내기")
dot.node("02-05-01-04", "비율을 백분율로 나타내기")

dot.node("02-06-01-01", "간단한 비례식 풀기")
dot.node("02-06-02-01", "주어진 양을 비례배분 하기")
```

```python
# 관계 설정
# 2. 변화와 관계

# 대분류 -> 중분류
dot.edge("02", "02-01")  # 변화와 관계 -> 규칙 찾기
dot.edge("02", "02-02")  # 변화와 관계 -> 규칙 정하기
dot.edge("02", "02-03")  # 변화와 관계 -> 등호와 동치 관계
dot.edge("02", "02-04")  # 변화와 관계 -> 대응 관계
dot.edge("02", "02-05")  # 변화와 관계 -> 비와 비율
dot.edge("02", "02-06")  # 변화와 관계 -> 비례식과 비례배분

# 중분류 -> 소분류
dot.edge("02-01", "02-01-01")  # 규칙 찾기 -> 배열에서 규칙 찾기
dot.edge("02-02", "02-02-01")  # 규칙 정하기 -> 규칙에 따라 배열하기
dot.edge("02-03", "02-03-01")  # 등호와 동치 관계 -> 등호 사용하기
dot.edge("02-04", "02-04-01")  # 대응 관계 -> 대응 관계를 나타낸 표에서 규칙 찾기
dot.edge("02-05", "02-05-01")  # 비와 비율 -> 두 양의 크기 비교하기
dot.edge("02-06", "02-06-01")  # 비례식과 비례배분 -> 비례식 이해하기
dot.edge("02-06", "02-06-02")  # 비례식과 비례배분 -> 비례배분 이해하기

# 소분류 -> 하위분류 (배열에서 규칙 찾기)
dot.edge(
    "02-01-01", "02-01-01-01"
)  # 배열에서 규칙 찾기 -> 물체 배열 규칙을 여러 가지 방법으로 표현하기
dot.edge(
    "02-01-01", "02-01-01-02"
)  # 배열에서 규칙 찾기 -> 무늬 배열 규칙을 여러 가지 방법으로 표현하기
dot.edge(
    "02-01-01", "02-01-01-03"
)  # 배열에서 규칙 찾기 -> 수 배열 규칙을 여러 가지 방법으로 표현하기
dot.edge(
    "02-01-01", "02-01-01-04"
)  # 배열에서 규칙 찾기 -> 다양한 변화 규칙을 수로 나타내기
dot.edge(
    "02-01-01", "02-01-01-05"
)  # 배열에서 규칙 찾기 -> 다양한 변화 규칙을 식으로 나타내기
dot.edge(
    "02-01-01", "02-01-01-06"
)  # 배열에서 규칙 찾기 -> 계산식 배열 규칙으로 계산 결과 추측하기

# 소분류 -> 하위분류 (규칙에 따라 배열하기)
dot.edge("02-02-01", "02-02-01-01")  # 규칙에 따라 배열하기 -> 물체 배열하기
dot.edge("02-02-01", "02-02-01-02")  # 규칙에 따라 배열하기 -> 무늬 배열하기
dot.edge("02-02-01", "02-02-01-03")  # 규칙에 따라 배열하기 -> 수 배열하기

# 엣지 추가
dot.edge("02-02-01-01", "02-01-01")
dot.edge("02-02-01-02", "02-01-01")
dot.edge("02-02-01-03", "02-01-01")

# 소분류 -> 하위분류 (등호 사용하기)
dot.edge(
    "02-03-01", "02-03-01-01"
)  # 등호 사용하기 -> 크기가 같은 두 양의 관계를 식으로 나타내기

# 소분류 -> 하위분류 (대응 관계를 나타낸 표에서 규칙 찾기)
dot.edge("02-04-01", "02-04-01-01")  # 대응 관계를 나타낸 표에서 규칙 찾기 -> □, △ 를 사용하여 식으로 나타내기

# 소분류 -> 하위분류 (두 양의 크기 비교하기)
dot.edge("02-05-01", "02-05-01-01")  # 두 양의 크기 비교하기 -> 두 양의 관계를 비로 나타내기
dot.edge("02-05-01", "02-05-01-02")  # 두 양의 크기 비교하기 -> 비율을 분수로 나타내기
dot.edge("02-05-01", "02-05-01-03")  # 두 양의 크기 비교하기 -> 비율을 소수로 나타내기
dot.edge("02-05-01", "02-05-01-04")  # 두 양의 크기 비교하기 -> 비율을 백분율로 나타내기

# 소분류 -> 하위분류 (비례식 이해하기)
dot.edge("02-06-01", "02-06-01-01")  # 비례식 이해하기 -> 간단한 비례식 풀기

# 소분류 -> 하위분류 (비례배분 이해하기)
dot.edge("02-06-02", "02-06-02-01")  # 비례배분 이해하기 -> 주어진 양을 비례배분 하기
```

```python
# 노드 추가
# 4. 자료와 가능성
dot.node("04", "자료와 가능성")  # 대분류


dot.node("04-01", "분류하기")  # 중분류
dot.node("04-02", "자료의 수집과 정리")
dot.node("04-03", "가능성")


dot.node("04-01-01", "개수 세기")  # 소분류
dot.node("04-01-02", "표로 나타내기")
dot.node("04-01-03", "○, ×, / 를 이용한 그래프로 나타내기")
dot.node("04-02-01", "그래프로 나타내기")
dot.node("04-02-02", "평균의 의미 알기")
dot.node("04-03-01", "사건이 일어날 가능성")


dot.node("04-01-01-01", "기준에 따른 결과 말하기")  # 하위분류
dot.node("04-01-02-01", "표로 나타내면 편리한 점 말하기")
dot.node("04-01-03-01", "그래프로 나타내면 편리한 점 말하기")

dot.node("04-02-01-01", "그림그래프로 나타내고 해석하기")
dot.node("04-02-01-02", "막대그래프로 나타내고 해석하기")
dot.node("04-02-01-03", "꺾은선그래프로 나타내고 해석하기")
dot.node("04-02-01-04", "띠그래프로 나타내고 해석하기")
dot.node("04-02-01-05", "원그래프로 나타내고 해석하기")

dot.node("04-02-02-01", "평균 구하고 해석하기")

dot.node("04-03-01-01", "가능성을 말로 표현하고 비교하기")
dot.node("04-03-01-02", "가능성을 수로 표현하기")
dot.node("04-03-01-03", "가능성을 예상하고 적절한 판단 내리기")
```

```python
# 관계 설정
# 4. 자료와 가능성

# 대분류 -> 중분류
dot.edge("04", "04-01")  # 자료와 가능성 -> 분류하기
dot.edge("04", "04-02")  # 자료와 가능성 -> 자료의 수집과 정리
dot.edge("04", "04-03")  # 자료와 가능성 -> 가능성

# 중분류 -> 소분류
dot.edge("04-01", "04-01-01")  # 분류하기 -> 개수 세기
dot.edge("04-01", "04-01-02")  # 분류하기 -> 표로 나타내기
dot.edge("04-01", "04-01-03")  # 분류하기 -> ○, ×, / 를 이용한 그래프로 나타내기
dot.edge("04-02", "04-02-01")  # 자료의 수집과 정리 -> 그래프로 나타내기
dot.edge("04-02", "04-02-02")  # 자료의 수집과 정리 -> 평균의 의미 알기
dot.edge("04-03", "04-03-01")  # 가능성 -> 사건이 일어날 가능성

# 소분류 -> 하위분류 (개수 세기)
dot.edge("04-01-01", "04-01-01-01")  # 개수 세기 -> 기준에 따른 결과 말하기
# 소분류 -> 하위분류 (표로 나타내기)
dot.edge("04-01-02", "04-01-02-01")  # 표로 나타내기 -> 표로 나타내면 편리한 점 말하기
# 소분류 -> 하위분류 (○, ×, / 를 이용한 그래프로 나타내기)
dot.edge("04-01-03", "04-01-03-01")  # ○, ×, / 를 이용한 그래프로 나타내기 -> 그래프로 나타내면 편리한 점 말하기

# 엣지 추가
dot.edge("04-01-03", "04-02-01-01")
dot.edge("04-01-03", "04-02-01-02")
dot.edge("04-01-03", "04-02-01-03")

# 소분류 -> 하위분류 (그래프로 나타내기)
dot.edge("04-02-01", "04-02-01-01")  # 그래프로 나타내기 -> 그림그래프로 나타내고 해석하기
dot.edge("04-02-01", "04-02-01-02")  # 그래프로 나타내기 -> 막대그래프로 나타내고 해석하기
dot.edge("04-02-01", "04-02-01-03")  # 그래프로 나타내기 -> 꺾은선그래프로 나타내고 해석하기
dot.edge("04-02-01", "04-02-01-04")  # 그래프로 나타내기 -> 띠그래프로 나타내고 해석하기
dot.edge("04-02-01", "04-02-01-05")  # 그래프로 나타내기 -> 원그래프로 나타내고 해석하기

# 소분류 -> 하위분류 (평균의 의미 알기)
dot.edge("04-02-02", "04-02-02-01")  # 평균의 의미 알기 -> 평균 구하고 해석하기

# 소분류 -> 하위분류 (사건이 일어날 가능성)
dot.edge("04-03-01", "04-03-01-01")  # 사건이 일어날 가능성 -> 가능성을 말로 표현하고 비교하기
dot.edge("04-03-01", "04-03-01-02")  # 사건이 일어날 가능성 -> 가능성을 수로 표현하기
dot.edge("04-03-01", "04-03-01-03")  # 사건이 일어날 가능성 -> 가능성을 예상하고 적절한 판단 내리기
```
