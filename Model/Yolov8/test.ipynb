# 패키지

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import os
from PIL import ImageDraw, Image
```

# 모델

```python
model = YOLO('yolov8n.pt')
```

```python
image_folder = "C:/Users/dongi/OneDrive/바탕 화면/kst_pipeline/Model/Yolov8/data/images"
output_folder = "C:/Users/dongi/OneDrive/바탕 화면/kst_pipeline/Model/Yolov8/data/images_test"
image_files = [os.path.join(image_folder, file) for file in os.listdir(image_folder) if file.endswith(('.png', '.jpg'))]
```

```python
os.makedirs(output_folder, exist_ok=True)
detections = []

for image_file in image_files:
    print("시작")
    results = model(image_file,conf = 0.25) # 여기서 민감도를 조절, 텍스트라는 점에서 기존보다 좀더 낮게 설정함
    detections.append(results)

    result_image = results[0].plot()
    output_path = os.path.join(output_folder, os.path.basename(image_file))
    plt.imsave(output_path, result_image)
    

for i, result in enumerate(detections):
    print(f"Image {i+1}:")
    print("Bounding Boxes (xyxy):", result[0].boxes.xyxy)  # 바운딩 박스 좌표
    print("Class IDs:", result[0].boxes.cls)  # 감지된 객체 클래스 ID
    print("Confidence Scores:", result[0].boxes.conf)  # 감지 신뢰도
```

```python
os.makedirs(output_folder, exist_ok=True)
detections = []

for image_file in image_files:
    print("시작")
    results = model(image_file) # 여기서 민감도를 조절, 텍스트라는 점에서 기존보다 좀더 낮게 설정함
    
    img = Image.open(image_file)
    draw = ImageDraw.Draw(img)

    text_boxes = []
    non_text_boxes = []

    for box,cls in zip(results[0].boxes.xyxy, results[0].boxes.cls):
        cls_id = int(cls)
        x1,y1,x2,y2 = map(int,box.tolist())

        if cls_id in [0, 1]:  # 텍스트로 간주할 클래스 ID 설정 (필요 시 수정)
            text_boxes.append((x1, y1, x2, y2))
            draw.rectangle((x1, y1, x2, y2), outline="red", width=2)  # 텍스트 영역 빨간색
        else:
            non_text_boxes.append((x1, y1, x2, y2))
            draw.rectangle((x1, y1, x2, y2), outline="blue", width=2)  # 비텍스트 영역 파란색

    
    
    
    

    # 결과 저장
    output_path = os.path.join(output_folder, os.path.basename(image_file))
    img.save(output_path)  # 이미지 저장
    print(f"Results saved to {output_path}")
    
    # 영역 출력
    print(f"Detected Text Boxes in {image_file}: {text_boxes}")
    print(f"Detected Non-Text Boxes in {image_file}: {non_text_boxes}")
```
